```python
import pandas as pd
```

```python
from sklearn.metrics import cohen_kappa_score
```

```python
#pd.read_csv("")
```

```python
def column_reconcile(file_name,header=[0,1],compared_columns=[1,3,8]):
    chl_pd = pd.read_csv(file_name,header=header)

    chl_pd_column_reconcile = pd.DataFrame([chl_pd.iloc[:,compared_columns[0]].str.strip().str.lower(),chl_pd.iloc[:,compared_columns[-2]].str.strip().str.lower(),chl_pd.iloc[:,compared_columns[-1]].str.strip().str.lower()]).transpose()

    chl_pd_column_reconcile = chl_pd_column_reconcile.fillna("NA")

    chl_pd_column_reconcile["agreement"] = chl_pd_column_reconcile.iloc[:,-2] == chl_pd_column_reconcile.iloc[:,-1]

    chl_pd_column_reconcile[chl_pd_column_reconcile["agreement"]==False]
    
    kappa_score = cohen_kappa_score(chl_pd_column_reconcile.iloc[:,-3].values,chl_pd_column_reconcile.iloc[:,-2].values)
    
    return chl_pd_column_reconcile,kappa_score,chl_pd_column_reconcile[chl_pd_column_reconcile["agreement"]==False]
```

```python
res1,res2,res3 = column_reconcile("Reconcile - BRA.csv",compared_columns=[1,2,4,9])
```
